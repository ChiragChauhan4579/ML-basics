# Central limit theorem

Consider a random variable X. This variable may or may not follow a gaussian distribution.

But if take random samples from this population data of X then the kde plot of the means of this samples follow a guassian distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr = np.random.randint(0,1000,500)
```

```python
sns.kdeplot(arr)
```

```python
s1 = np.random.choice(arr,100)
s2 = np.random.choice(arr,100)
s3 = np.random.choice(arr,100)
s4 = np.random.choice(arr,100)
s5 = np.random.choice(arr,100)
```

```python
print(np.mean(s1))
print(np.mean(s2))
print(np.mean(s3))
print(np.mean(s4))
print(np.mean(s5))
```

```python
lis = [np.mean(s1),np.mean(s2),np.mean(s3),np.mean(s4),np.mean(s5)]
```

```python
new_arr = np.array(lis)
```

```python
sns.kdeplot(new_arr)
```
